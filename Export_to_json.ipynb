```
# importing required modules
import PyPDF2
  
# creating a pdf file object
pdfFileObj = open('publicregisterregularlabourandhighlyskilledmigrants.pdf', 'rb')
  
# creating a pdf reader object
pdfReader = PyPDF2.PdfReader(pdfFileObj)
```

```
num_of_pages=len(pdfReader.pages)
```

```
print(num_of_pages)
```

```
def check_company(str):
  str=str.strip()
  if len(str)==0: return False #empty str
  if str.startswith("Pagina"): return False #page
  if not str.split()[-1].isdigit(): return False #not a company without numbers
  return True
  
```

```
print('Pagina 1 van 185 ', check_company('Pagina 1 van 185 '))
print(' Erkende referenten / Recognised sponsors  ', check_company(' Erkende referenten / Recognised sponsors  '))
print('PTW Benelux B.V.  24395334  ', check_company('PTW Benelux B.V.  24395334  '))
print(" ", check_company(' '))
```

```
pdfReader.pages[0].extract_text().split('\n')
```

```
print( list( filter(check_company, pdfReader.pages[0].extract_text().split('\n')) ) )
```

```
companies_lsts=[]
for i in range(num_of_pages):
  new_str_lists=list( filter(check_company, pdfReader.pages[i].extract_text().split('\n'))  )
  new_company_lists= list( map( lambda x: ''.join(x.split()[:-1]), new_str_lists) )
  companies_lsts+=new_company_lists
```

```
companies_lsts
```

```
import json

data = {"sponsors": companies_lsts}

# Convert the dictionary to a JSON string
json_data = json.dumps(data)

# Save the JSON string to a file
with open("km_company.json", "w") as file:
    file.write(json_data)
```
